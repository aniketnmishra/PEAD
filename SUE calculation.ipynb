```python
import pandas as pd
import numpy as np
```

```python
actual_eps = pd.read_csv('eps_nifty50.csv')
eps_pred = pd.read_csv('linear_reg_predicted_eps.csv')
std_dev = pd.read_csv('std_dev_eps_predicted.csv')
```

```python
actual_eps.drop(index=list(range(48, 997)),inplace=True)
actual_eps.drop('Company',inplace=True,axis=1)
```

```python
Ticker_list = eps_pred['Unnamed: 0']
```

```python
eps_pred.rename(columns={'Unnamed: 0':'Ticker'},inplace=True)
std_dev.rename(columns={'Unnamed: 0':'Ticker'},inplace=True)
```

```python
print(actual_eps.shape)
print(eps_pred.shape)
print(std_dev.shape)
```

```python
to_be_removed = actual_eps.columns.tolist()[31:]
actual_eps.drop(to_be_removed,axis=1,inplace=True)
```

```python
print(actual_eps.shape)
print(eps_pred.shape)
print(std_dev.shape)
```

```python
actual_eps.drop('Ticker',inplace=True,axis=1)
eps_pred.drop('Ticker',inplace=True,axis=1)
std_dev.drop('Ticker',inplace=True,axis=1)
```

```python
unexpected_earning = actual_eps-eps_pred
std_dev.head()
```

```python
sue = unexpected_earning.truediv(std_dev)
```

```python
sue['Ticker']=Ticker_list
```

```python
sue.set_index('Ticker',inplace=True)
```

```python
sue.reset_index(inplace=True)
```

```python
sue.head()
```

```python
sue.to_csv(r'sue.csv')
```

```python
cutoffs = {}
for date in sue.columns.tolist()[2:]:
    cutoff_list = []
    cutoff_list.append(np.percentile(sue[date],10))    
    cutoff_list.append(np.percentile(sue[date],15))
    cutoff_list.append(np.percentile(sue[date],20))
    cutoff_list.append(np.percentile(sue[date],25))
    cutoff_list.append(np.percentile(sue[date],30))
    cutoff_list.append(np.percentile(sue[date],40))
    cutoff_list.append(np.percentile(sue[date],50))
    cutoff_list.append(np.percentile(sue[date],60))
    cutoff_list.append(np.percentile(sue[date],70))
    cutoff_list.append(np.percentile(sue[date],75))
    cutoff_list.append(np.percentile(sue[date],80))
    cutoff_list.append(np.percentile(sue[date],85))
    cutoff_list.append(np.percentile(sue[date],90))
    cutoffs[date] = cutoff_list
column_list = [10,15,20,25,30,40,50,60,70,75,80,85,90]
cutoff = pd.DataFrame.from_dict(cutoffs, orient='index',columns=column_list)
#cutoff
cutoff.to_csv(r'sue_cutoffs.csv')
```
