```python
#from pandas import read_csv
from pandas import datetime
import pandas as pd
import numpy as np
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

reading in eps data

```python
data = pd.read_csv('eps_nifty50.csv')
data
```

removing extra NaN rows that came in the datframe while reading from csv

```python
data.drop(index=list(range(48, 997)),inplace=True)
data.drop('Company',inplace=True,axis=1)
```

making a list of column names

```python
column_name = data.columns[1:31].tolist()
```

predicting eps using linear regression

```python
pred_dict = {}
std_dict = {}
for i in range(len(data)):
    prediction=[]
    stddev=[]
    for j in range(1,31):
        series = data.iloc[i][(j+1):(j+17)]
        X = np.asarray([i for i in range(16)]).reshape(-1,1)
        y = series.copy()
        X_pred =np.asarray([16]).reshape(-1,1)
        regressor = LinearRegression()  
        regressor.fit(X, y)
        y_pred = regressor.predict(X_pred)
        std = np.std(series)
        prediction.append(y_pred[0])
        stddev.append(std)
    pred_dict[data.loc[i][0]] = prediction
    std_dict[data.loc[i][0]] = stddev
pred = pd.DataFrame.from_dict(pred_dict, orient='index',columns=column_name)
stdev = pd.DataFrame.from_dict(std_dict, orient='index',columns=column_name)
#pred.head()
```

saving dataframes to csv

```python
pred.to_csv(r'linear_reg_predicted_eps.csv')
```

```python
stdev.to_csv(r'std_dev_eps_predicted.csv')
```
